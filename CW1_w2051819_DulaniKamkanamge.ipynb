#### **1) Multiple Linear Regression**

**Overview:**

Multiple linear regression uses a linear equation to model the connection between several input parameters and a continuous target variable.  It acts as a straightforward starting point for using structured features to forecast traffic volume.

**Strengths & Advantages:**

**•** Computational Efficiency: Very quick training and prediction, appropriate for real-time applications.

**•** High Interpretability: Coefficients show feature importance (e.g., rush hour impact, weather affects).

**•** Low Overfitting Risk: Generalises well to unseen data with appropriate regularisation (Ridge/Lasso).

**•** Handles Mixed Features: Can incorporate both continuous (temperature) and encoded categorical (weather) variables.

**•** Statistical Diagnostics: Offers p-values, confidence intervals, and R² metrics for model validation.


**Weaknesses & Disadvantages:**

**•** No Temporal Memory: Treats each hour separately, disregarding sequence dependencies.

**•** Linearity Assumption: Unable to grasp intricate nonlinear linkages in traffic patterns.

**•** Feature Engineering Intensive: Manual feature creation is a major factor in performance.

**•** Multicollinearity Problems: Coefficients may be distorted by correlated features (hour and rush hour indicator).

**•** Outlier Sensitivity: Model fit is greatly impacted by high traffic volumes.


**Input Data & Expected Output:**

**Input:** Feature matrix with engineered variables:

**•** Temporal: hour (cyclically encoded), day_of_week, is_weekend, is_holiday

**•** Weather: temperature, rain_1h, snow_1h, clouds_all, encoded weather_main

**•** Historical: lagged traffic_volume (t-1, t-24, t-168)

**•** Derived: is_rush_hour, is_night, season


**Output:** Predicted traffic_volume (continuous integer) for a specific future hour.

**Application to Traffic Prediction:**

Linear regression can establish baseline relationships between observable conditions and traffic volume.  For instance, it might discover that traffic is always 15% higher on Friday afternoons when the weather is clear than on Tuesday mornings when it rains.  Nevertheless, it is unable to simulate abrupt shifts from events or intricate temporal dynamics.

**Suitability Assessment:**

Moderately appropriate for baseline comparison and preliminary investigation.  gives understandable insights, but because of nonlinear traffic dynamics, it is not sophisticated enough for precise real-world prediction.


#### **2) Long Short-Term Memory (LSTM) Network**

**Overview:**

Designed for sequential data, LSTM is a specialised recurrent neural network architecture that preserves memory cells that capture long-term dependencies, making it perfect for time-series traffic prediction.

**Strengths & Advantages:**

**•** Temporal Dependency Modeling: Naturally captures daily, weekly, and seasonal patterns in traffic.

**•** Non-linear Pattern Recognition: Learns complex relationships without extensive feature engineering.

**•** Sequence-to-Sequence Capability: Can predict multiple future time steps from historical sequences.

**•** State-of-the-Art Performance: Consistently outperforms traditional methods on time-series benchmarks.

**•** Multivariate Input Support: Can process traffic volume alongside weather features in sequence format.

**•** Memory Mechanisms: Explicitly models both short-term fluctuations and long-term trends.

**Weaknesses & Disadvantages:**

**•** Computational Intensity: Requires significant training time and resources (GPUs recommended).

**•** Black-Box Nature: Difficult to interpret predictions and understand decision logic.

**•** Data Requirements: Needs substantial historical data (thousands of samples) for effective training.

**•** Hyperparameter Sensitivity: Performance depends on careful tuning of layers, units, dropout rates.

**•** Training Instability: Can suffer from vanishing/exploding gradients without proper initialization.

**•** Sequence Length Dependency: Fixed input sequence length may not capture all relevant patterns.


**Input Data & Expected Output:**

**Input:** Sequential matrices of shape (samples, timesteps, features):

**•** timesteps = 24 (previous 24 hours)

**•** features = [traffic_volume, temp, rain_1h, clouds_all, encoded_time_features]

Each sample represents 24 consecutive hours of multi-feature observations

**Output:** Predicted traffic_volume for the next hour (single-step) or next N hours (multi-step).

**Application to Traffic Prediction:**

LSTM can discover that on weekdays, traffic usually increases between 6 and 9 AM, decreases over lunch, peaks again between 5 and 7 PM, and exhibits distinct patterns on weekends and holidays. It can also learn how the traffic flow throughout the day is impacted by rainfall at 8 AM.

**Suitability Assessment:**

High suitability for the given objective. provides precise multi-step forecasts that are necessary for proactive traffic management on the Metro Interstate and directly handles the sequential nature of traffic data.



#### **3) Random Forest Regression**

**Overview:**

For structured data with a variety of feature types, Random Forest is an ensemble learning technique that builds several decision trees and aggregates their predictions.

**Strengths & Advantages:**

**•** Handles Mixed Data Types: Naturally supports both continuous (temperature) and categorical (weather) features.

**•** Feature Importance Rankings: Offers precise metrics on which variables most affect traffic volume.

**•** Robust to Outliers: Less susceptible to extreme values than linear methods.

**•** Non-linear Modelling: Captures intricate interactions without explicit specification.

**•** Reduced Overfitting: Ensemble approach with bagging offers good generalisation.

**•** Parallelizable: For efficiency, training can be divided over several cores.

**Weaknesses & Disadvantages:**

**•** Memory-intensive: Stores all trees, necessitating substantial memory for large ensembles.

**•** No Temporal Awareness: Treats each time point independently, disregarding sequence context.

**•** Limited Extrapolation: Doesn't perform well outside of the training data range.

**•** Longer Inference Time: Compared to linear models, multiple tree traversals slow prediction.

**•** Black-Box Elements: Although feature importance is known, individual predictions are opaque.

**•** Hyperparameter Tuning: Careful optimisation is needed for the number of trees, depth, and sample splitting.

**Input Data & Expected Output:**

**Input:** Feature vector for a specific hour:

**•** Raw features: hour, temp, rain_1h, snow_1h, clouds_all

**•** Engineered features: is_holiday, day_of_week_encoded, is_rush_hour

**•** Lag features: traffic_volume_prev_hour, traffic_volume_same_hour_yesterday

**•** Rolling statistics: avg_traffic_last_3_hours, avg_traffic_last_24_hours

**Output:** Predicted traffic_volume with uncertainty estimates from tree variance.

**Application to Traffic Prediction:**

The combination of "Friday + 5 PM + Rain + Temperature between 280-285K" usually results in 20% more traffic than baseline, according to Random Forest. Although it is excellent at capturing intricate feature interactions, it is unable to capture the temporal evolution of congestion.

**Suitability Assessment:**

Good suitability for static prediction and feature analysis. Excellent benchmark performance and comprehensible insights are provided, but careful feature engineering is needed to make up for the absence of temporal modelling.

### 4. Advanced Feature Engineering

```python
# Create comprehensive temporal and cyclical features
def create_advanced_features(df):
    data = df.copy()

    # Convert date_time to datetime
    data['date_time'] = pd.to_datetime(data['date_time'])

    # Extract comprehensive temporal features
    data['year'] = data['date_time'].dt.year
    data['month'] = data['date_time'].dt.month
    data['day'] = data['date_time'].dt.day
    data['hour'] = data['date_time'].dt.hour
    data['day_of_week'] = data['date_time'].dt.dayofweek
    data['day_of_year'] = data['date_time'].dt.dayofyear
    data['week_of_year'] = data['date_time'].dt.isocalendar().week
    data['quarter'] = data['date_time'].dt.quarter
    data['is_weekend'] = (data['day_of_week'] >= 5).astype(int)

    # Cyclical encoding for hour, day, month
    data['hour_sin'] = np.sin(2 * np.pi * data['hour'] / 24)
    data['hour_cos'] = np.cos(2 * np.pi * data['hour'] / 24)
    data['day_sin'] = np.sin(2 * np.pi * data['day'] / 31)
    data['day_cos'] = np.cos(2 * np.pi * data['day'] / 31)
    data['month_sin'] = np.sin(2 * np.pi * data['month'] / 12)
    data['month_cos'] = np.cos(2 * np.pi * data['month'] / 12)
    data['day_of_week_sin'] = np.sin(2 * np.pi * data['day_of_week'] / 7)
    data['day_of_week_cos'] = np.cos(2 * np.pi * data['day_of_week'] / 7)

    # Create lag features
    data['traffic_lag_1'] = data['traffic_volume'].shift(1)
    data['traffic_lag_24'] = data['traffic_volume'].shift(24)  # Same hour previous day
    data['traffic_lag_168'] = data['traffic_volume'].shift(168)  # Same hour previous week

    # Rolling statistics
    data['traffic_rolling_3h_mean'] = data['traffic_volume'].rolling(window=3, min_periods=1).mean()
    data['traffic_rolling_6h_mean'] = data['traffic_volume'].rolling(window=6, min_periods=1).mean()
    data['traffic_rolling_24h_mean'] = data['traffic_volume'].rolling(window=24, min_periods=1).mean()
    data['traffic_rolling_3h_std'] = data['traffic_volume'].rolling(window=3, min_periods=1).std()

    # Time-based features
    data['is_rush_hour_morning'] = ((data['hour'] >= 7) & (data['hour'] <= 9)).astype(int)
    data['is_rush_hour_evening'] = ((data['hour'] >= 16) & (data['hour'] <= 18)).astype(int)
    data['is_night'] = ((data['hour'] >= 22) | (data['hour'] <= 5)).astype(int)
    data['is_work_hours'] = ((data['hour'] >= 8) & (data['hour'] <= 17)).astype(int)

    # Holiday features
    data['is_holiday'] = data['holiday'].notnull().astype(int)
    data['holiday_tomorrow'] = data['holiday'].shift(-24).notnull().astype(int)
    data['holiday_yesterday'] = data['holiday'].shift(24).notnull().astype(int)

    # Weather encoding
    le = LabelEncoder()
    data['weather_main_encoded'] = le.fit_transform(data['weather_main'])

    # Temperature in Celsius (from Kelvin)
    data['temp_celsius'] = data['temp'] - 273.15

    # Weather severity score
    weather_severity = {
        'Clear': 1, 'Clouds': 2, 'Haze': 3, 'Mist': 3,
        'Fog': 4, 'Drizzle': 4, 'Rain': 5, 'Thunderstorm': 6,
        'Snow': 7
    }
    data['weather_severity'] = data['weather_main'].map(weather_severity).fillna(3)

    # Create interaction features
    data['temp_times_hour'] = data['temp'] * data['hour']
    data['clouds_times_hour'] = data['clouds_all'] * data['hour']

    # Drop rows with NaN from lag features
    data = data.dropna()

    return data
```

### 5. Data Preprocessing and Feature Selection

```python
# Prepare data for LSTM training with feature selection
def prepare_data_for_lstm(df, target_col='traffic_volume', seq_length=24, test_size=0.2):
    data = df.copy()

    # Select features for modeling
    feature_cols = [
        # Original features
        'temp', 'rain_1h', 'snow_1h', 'clouds_all',

        # Engineered temporal features
        'hour_sin', 'hour_cos', 'day_of_week_sin', 'day_of_week_cos',
        'month_sin', 'month_cos',

        # Lag features
        'traffic_lag_1', 'traffic_lag_24', 'traffic_lag_168',

        # Rolling statistics
        'traffic_rolling_3h_mean', 'traffic_rolling_24h_mean',

        # Time indicators
        'is_weekend', 'is_rush_hour_morning', 'is_rush_hour_evening',
        'is_night', 'is_work_hours',

        # Holiday features
        'is_holiday', 'holiday_tomorrow',

        # Weather features
        'weather_severity', 'weather_main_encoded',

        # Interaction features
        'temp_times_hour', 'clouds_times_hour'
    ]

    # Ensure all columns exist
    feature_cols = [col for col in feature_cols if col in data.columns]

    # Create feature matrix and target
    X = data[feature_cols].values
    y = data[target_col].values

    print(f"Selected {len(feature_cols)} features:")
    print(feature_cols)

    # Normalize features
    scaler_X = StandardScaler()
    X_scaled = scaler_X.fit_transform(X)

    # Normalize target
    scaler_y = StandardScaler()
    y_scaled = scaler_y.fit_transform(y.reshape(-1, 1)).flatten()

    # Create sequences for LSTM
    def create_sequences(X, y, seq_length):
        X_seq, y_seq = [], []
        for i in range(len(X) - seq_length):
            X_seq.append(X[i:(i + seq_length)])
            y_seq.append(y[i + seq_length])
        return np.array(X_seq), np.array(y_seq)

    X_seq, y_seq = create_sequences(X_scaled, y_scaled, seq_length)

    # Split data chronologically
    split_idx = int(len(X_seq) * (1 - test_size))

    X_train = X_seq[:split_idx]
    X_test = X_seq[split_idx:]
    y_train = y_seq[:split_idx]
    y_test = y_seq[split_idx:]

    print(f"\nData Preparation Summary:")
    print(f"Total sequences: {len(X_seq)}")
    print(f"Training samples: {X_train.shape[0]}")
    print(f"Testing samples: {X_test.shape[0]}")
    print(f"Sequence shape: {X_train.shape[1:]} (timesteps, features)")

    return X_train, X_test, y_train, y_test, scaler_X, scaler_y, feature_cols, data
```

```python
# Prepare data
SEQ_LENGTH = 24  # Use 24 hours of historical data
X_train, X_test, y_train, y_test, scaler_X, scaler_y, feature_cols, processed_data = prepare_data_for_lstm(
    df_engineered, seq_length=SEQ_LENGTH, test_size=0.2
)
```

### 6. Build Enhanced LSTM Model

```python
# Build an enhanced LSTM model with multiple layers and regularization
def build_enhanced_lstm_model(input_shape, dropout_rate=0.3, lstm_units=100):
    model = Sequential([
        Input(shape=input_shape),

        # First LSTM layer with return sequences
        Bidirectional(LSTM(units=lstm_units,
                          return_sequences=True,
                          kernel_regularizer=l2(0.001),
                          recurrent_regularizer=l2(0.001))),
        Dropout(dropout_rate),

        # Second LSTM layer
        Bidirectional(LSTM(units=int(lstm_units/2),
                          return_sequences=True,
                          kernel_regularizer=l2(0.001))),
        Dropout(dropout_rate),

        # Third LSTM layer
        LSTM(units=int(lstm_units/4),
             kernel_regularizer=l2(0.001)),
        Dropout(dropout_rate),

        # Dense layers
        Dense(units=64, activation='relu', kernel_regularizer=l2(0.001)),
        Dropout(dropout_rate/2),

        Dense(units=32, activation='relu', kernel_regularizer=l2(0.001)),
        Dropout(dropout_rate/2),

        # Output layer
        Dense(units=1)
    ])

    # Compile with customized optimizer
    optimizer = Adam(learning_rate=0.001, clipnorm=1.0)

    # Try different loss functions. Use 'huber' instead of 'huber_loss'
    model.compile(
        optimizer=optimizer,
        loss='huber',  # Changed from 'huber_loss' to 'huber'
        metrics=['mae', 'mse']
    )

    # Display model summary
    model.summary()

    return model
```

```python
# Build model
input_shape = (X_train.shape[1], X_train.shape[2])
model = build_enhanced_lstm_model(input_shape)
```

```python
# Define advanced callbacks
callbacks = [
    EarlyStopping(
        monitor='val_loss',
        patience=15,
        restore_best_weights=True,
        verbose=1,
        min_delta=0.001
    ),
    ModelCheckpoint(
        'best_lstm_traffic_model_enhanced.keras',
        monitor='val_loss',
        save_best_only=True,
        verbose=0
    ),
    ReduceLROnPlateau(
        monitor='val_loss',
        factor=0.5,
        patience=7,
        min_lr=0.00001,
        verbose=1
    )
]
```

```python
# Calculate and display comprehensive regression metrics
def calculate_comprehensive_metrics(y_true, y_pred):
    metrics = {
        'Mean Absolute Error (MAE)': mean_absolute_error(y_true, y_pred),
        'Root Mean Squared Error (RMSE)': np.sqrt(mean_squared_error(y_true, y_pred)),
        'Mean Absolute Percentage Error (MAPE)': mean_absolute_percentage_error(y_true, y_pred) * 100,
        'R-squared (R²)': r2_score(y_true, y_pred),
        'Explained Variance Score': 1 - (np.var(y_true - y_pred) / np.var(y_true))
    }

    metrics_df = pd.DataFrame({
        'Metric': list(metrics.keys()),
        'Value': list(metrics.values()),
        'Interpretation': [
            'Average absolute error',
            'Root mean squared error (penalizes large errors)',
            'Average percentage error',
            'Variance explained by model',
            'Proportion of variance explained'
        ]
    })

    return metrics_df, metrics

metrics_df, metrics = calculate_comprehensive_metrics(y_test_actual, y_pred_actual)
print("MODEL PERFORMANCE METRICS")
print("="*60)
print(metrics_df.to_string(index=False))
print("\n" + "="*60)
print("INTERPRETATION:")
print(f"- Predictions are off by {metrics['Mean Absolute Error (MAE)']:.0f} vehicles on average")
print(f"- MAPE of {metrics['Mean Absolute Percentage Error (MAPE)']:.1f}% indicates good relative accuracy")
print(f"- Model explains {metrics['R-squared (R²)']:.2%} of variance in traffic volume")
print(f"- RMSE of {metrics['Root Mean Squared Error (RMSE)']:.0f} shows prediction spread")
```

```python
# Generate multi-step forecast with feature propagation
def forecast_multistep(model, last_sequence, steps=24, scaler_y=scaler_y, scaler_X=scaler_X,
                      feature_cols=feature_cols, processed_data=processed_data):
    forecasts = []
    confidence_intervals = []
    current_sequence = last_sequence.copy()

    # Get the last row of data for feature updates
    last_row = processed_data.iloc[-1:].copy()

    for step in range(steps):
        # Reshape for prediction
        current_input = current_sequence.reshape(1, SEQ_LENGTH, len(feature_cols))

        # Predict next step
        next_step_scaled = model.predict(current_input, verbose=0)[0, 0]
        forecasts.append(next_step_scaled)

        # Create new feature vector for next timestep
        new_features = current_sequence[-1].copy()

        # Update temporal features for next hour
        next_hour = (pd.to_datetime(last_row['date_time'].iloc[0]) +
                    timedelta(hours=step+1)).hour
        next_day_of_week = (pd.to_datetime(last_row['date_time'].iloc[0]) +
                           timedelta(hours=step+1)).dayofweek

        # Update cyclical features
        new_features[feature_cols.index('hour_sin')] = np.sin(2 * np.pi * next_hour / 24)
        new_features[feature_cols.index('hour_cos')] = np.cos(2 * np.pi * next_hour / 24)
        new_features[feature_cols.index('day_of_week_sin')] = np.sin(2 * np.pi * next_day_of_week / 7)
        new_features[feature_cols.index('day_of_week_cos')] = np.cos(2 * np.pi * next_day_of_week / 7)

        # Update lag features (use predicted value)
        if 'traffic_lag_1' in feature_cols:
            new_features[feature_cols.index('traffic_lag_1')] = next_step_scaled

        # Update rolling statistics (simplified)
        if 'traffic_rolling_3h_mean' in feature_cols:
            # Use recent predictions for rolling mean
            recent_values = list(current_sequence[:, feature_cols.index('traffic_lag_1')][-2:]) + [next_step_scaled]
            new_features[feature_cols.index('traffic_rolling_3h_mean')] = np.mean(recent_values)

        # Update time indicators
        if 'is_rush_hour_morning' in feature_cols:
            new_features[feature_cols.index('is_rush_hour_morning')] = 1 if 7 <= next_hour <= 9 else 0
        if 'is_rush_hour_evening' in feature_cols:
            new_features[feature_cols.index('is_rush_hour_evening')] = 1 if 16 <= next_hour <= 18 else 0
        if 'is_night' in feature_cols:
            new_features[feature_cols.index('is_night')] = 1 if next_hour >= 22 or next_hour <= 5 else 0

        # Update sequence (remove first, append new features)
        current_sequence = np.roll(current_sequence, -1, axis=0)
        current_sequence[-1] = new_features

    # Convert back to original scale
    forecasts = np.array(forecasts).reshape(-1, 1)
    forecasts_actual = scaler_y.inverse_transform(forecasts).flatten()

    return forecasts_actual
```

```python
# Generate what-if scenarios for different conditions
def what_if_scenario(base_forecast, scenario_type='rain_storm', severity=1.0):
    scenarios = {
        'rain_storm': {'traffic_change': -0.15, 'description': 'Heavy Rain Storm'},
        'snow_emergency': {'traffic_change': -0.40, 'description': 'Snow Emergency'},
        'holiday_weekend': {'traffic_change': -0.20, 'description': 'Holiday Weekend'},
        'construction_closure': {'traffic_change': -0.30, 'description': 'Major Construction'},
        'special_event': {'traffic_change': 0.25, 'description': 'Major Event'},
        'clear_weather': {'traffic_change': 0.10, 'description': 'Perfect Clear Weather'}
    }

    if scenario_type not in scenarios:
        return base_forecast, "Normal Conditions"

    scenario = scenarios[scenario_type]
    adjusted_forecast = base_forecast * (1 + scenario['traffic_change'] * severity)

    return adjusted_forecast, scenario['description']

# Generate different scenarios
scenarios = ['rain_storm', 'snow_emergency', 'holiday_weekend', 'special_event', 'clear_weather']
scenario_forecasts = {}

for scenario in scenarios:
    adjusted_forecast, description = what_if_scenario(future_forecast, scenario)
    scenario_forecasts[description] = adjusted_forecast
```

Because of this, the model might make more accurate forecasts for locations that are closely watched, which could exacerbate already-existing disparities in resource allocation and traffic planning.

Furthermore, earlier policy choices that gave some commuter routes priority over others may be reflected in historical congestion patterns. By forecasting and optimising for historically prioritised traffic flows rather than equitable transport outcomes, the model may inadvertently reinforce such biases if they exist in the data.

**Accessibility and Inclusion**

The AI system's outputs, like traffic forecasts and visualisations, are intended to assist planners and traffic authorities. From an inclusion standpoint, the system might be expanded to accommodate many user groups, including the general public, public transportation providers, and municipal governments. The application of AI to traffic management raises ethical concerns concerning responsibility and transparency. ### **2. Environmental Sustainability**

**Computational Resource Analysis**

Because of their recurrent design and training procedure, LSTM models demand more processing power than more straightforward statistical or regression-based methods. Multiple epochs and matrix operations are required for model training, which raises energy consumption, especially when scaling to larger datasets or real-time prediction systems.

However, this method keeps computational costs within practical bounds for an academic prototype by training the model on a moderately big dataset using ordinary hardware.

**Data Storage Requirements**

The system depends on past traffic data, which can increase dramatically over time, particularly when it is gathered at a high temporal precision. Large amounts of raw data and learnt model checkpoints must be stored, which raises storage needs and related environmental expenses.

Data retention guidelines like the following could be used to lessen this:

**•** Reducing temporal resolution by combining older data

**•** Eliminating unnecessary or low-value features

**•** Keeping only the most current or pertinent training samples

**Reducing Environmental Impact in Production Deployment**

A production-level system's environmental impact could be decreased via a number of tactics:

**•** Model simplification: When applicable, swap out LSTM models for lighter architectures (like GRU or regression models).

**•** Edge or batch processing: To lessen computational effort, predictions are made in batches rather than continuously.
